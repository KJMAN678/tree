```python
from sklearn.datasets import load_iris
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
clf = tree.DecisionTreeClassifier(random_state=42)
iris = load_iris()
```

```python
clf = clf.fit(iris.data, iris.target)
plt.figure(figsize=(12, 8))
tree.plot_tree(clf)

plt.show()
```
